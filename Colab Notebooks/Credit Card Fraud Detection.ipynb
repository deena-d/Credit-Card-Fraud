# unsampling



where are taking random 492 columns in legit because legit have more than 2lacs data , it provide unstable prediction of data


```python
legit_sample = legit.sample(n=492)
```

concat two dataset


```python
new_df = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df.shape
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
x = new_df.drop(['Class','V11','V12','V13','V14','V15','V16','V17','V18','V19','V20','V21','V22','V23','V24','V25','V26','V27','V28'],axis=1)
y = new_df['Class']
```

```python
print(x)
```

```python
print(y)
```

split data into test and train data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Model Training

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

Model Evalution

Accuracy score

```python
x_train_prediction= model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy of the train data:",training_data_accuracy)
```

```python
x_test_prediction= model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy of the train data:",training_data_accuracy)
```
